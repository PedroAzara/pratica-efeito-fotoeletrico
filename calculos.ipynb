# Parte 1

Obtenção da constante de Planck usando o aparato fotoelétrico

```python
valores_lambda = np.array([365,405,436,546,577])*10**(-9)
valores_V = np.array([1.839,1.473,1.224,0.634, 0.574])
```

```python
import matplotlib.pyplot as plt

# Velocidade da luz em m/s
c = 3 * 10**8

# Calcular a frequência
valores_frequencia = c / valores_lambda

# Plotar o gráfico
plt.scatter(valores_frequencia, valores_V)
plt.xlabel('Frequência (Hz)')
plt.ylabel('Tensão (V)')
plt.title('Tensão em função da Frequência')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import linregress
carga_eletron = 1.6 * 10**(-19)

# Realizar a regressão linear
slope, intercept, r_value, p_value, std_err = linregress(valores_frequencia, valores_V)

# Gerar os valores ajustados
valores_ajustados = slope * valores_frequencia + intercept

# Plotar o gráfico com a regressão linear
plt.plot(valores_frequencia, valores_V, 'o', label='Dados experimentais')
plt.plot(valores_frequencia, valores_ajustados, 'r', label='Regressão linear')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Tensão (V)')
plt.title('Tensão em função da Frequência')
plt.legend()
plt.grid(True)
plt.show()
# Erro para a e b
erro_a = std_err
erro_b = std_err * np.sqrt(np.sum(valores_frequencia**2) / len(valores_frequencia))


# Valores de a e b
a = slope
b = intercept
print(f"Valor de h: {a*carga_eletron} +- {erro_a*carga_eletron}")
print(f"Valor de b: {b*carga_eletron} +- {erro_b*carga_eletron}")
```
